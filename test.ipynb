```python
%pip install pandas numpy matplotlib seaborn

import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns   
```

```python
salary_data = pd.read_csv(r"C:\Users\vkirt\Downloads\Salary_Data.csv")
salary_data
```

```python
x = salary_data[["YearsExperience"]] #Input (always provide inputs in list format)
y = salary_data["Salary"] #Output or Traget Variable
```

```python
x
```

```python
y
```

```python
#spiliting data into training and testing
# x-train input data for training
# x-test input data for testing
# y-train output data for training
# y-test output data for testing

# test_size = 0.3 means 30% of the data will be used for testing and 70% will be used for training
# random_state = 42 is the seed value, it is used to shuffle the data and get the same result every time
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
x_train.shape
```

```python
# select the model
from sklearn.linear_model import LinearRegression
mymodel = LinearRegression()
```

```python
# train the model
mymodel.fit(x_train, y_train)
```

```python
x_test
```

```python
# test the model
y_pred = mymodel.predict(x_test)
y_pred
```

```python
# evaluate the model
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
appliance_data = pd.read_csv(r"C:\Users\vkirt\Downloads\appliance_energy.csv")
appliance_data
```

```python
appliance_data.columns
```

```python
x = appliance_data[["Temperature (°C)"]]
y = appliance_data["Energy Consumption (kWh)"]
x
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
EnergyLRmodel = LinearRegression()
```

```python
EnergyLRmodel.fit(x_train, y_train)
```

```python
y_pred = EnergyLRmodel.predict(x_test)
```

```python
y_pred


```

```python
x_test
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
print(f"R2 Score: {r2}")
print(f"Mean Squared Error: {mse}")
print(f"Mean Absolute Error: {mae}")
```
